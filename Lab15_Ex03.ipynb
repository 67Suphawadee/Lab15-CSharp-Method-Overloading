## Runtime polymorphism

1.สร้าง console application project

```cmd
dotnet new console --name Lab15_Ex03
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var animalA = new Dog();
var animalB = new Fish();
var animalC = new Bird();

animalA.Move();
animalB.Move();
animalC.Move();

class Animal
{
   public virtual void Move()
   {
    WriteLine("Move successfully.");
   } 
}

class Dog:Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Running on the ground.");
        base.Move();
    }
}
class Fish: Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Swimming in the water.");
        base.Move();
    }
}

class Bird: Animal
{
    public override void Move()
    {
        WriteLine($"{this.GetType()}: Flying in the air.");
        base.Move();
    }
}
```

Move() ถูก overridden ในคลาสที่สืบทอด (เช่น Dog, Fish, Bird) ทำให้เราสามารถเรียก method เดียวกันจากอ็อบเจ็กต์ต่างชนิดกันได้ (ในนี้คืออ็อบเจ็กต์ของ Dog, Fish, และ Bird) แต่ผลลัพธ์ที่ได้จะแตกต่างกันตามชนิดของอ็อบเจ็กต์นั้นๆ
